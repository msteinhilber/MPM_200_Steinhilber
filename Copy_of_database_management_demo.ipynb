## Population of states (summarizing the data)

```python
states = pd.read_excel("NST-EST2023-POP.xlsx")
```

```python
states.head()
```

```python
# name the columns
states.columns = ['state', '2024', '2020', '2021', '2022', '2023']
states.head()
```

```python
states.state.unique().tolist() #need to fix the issue of states being named differently than what we refer to
```

```python
states['state'] = states['state'].map(lambda x: x.lstrip('.'))
states.state.unique().tolist()
# cleans up data in terms of how states are named
# to clean up data, search in google or chatgpt: regular expression (regex) code for cleaning of "whatever you want to remove"
# regular expression is keywords to work with text data (used for finding and replacing substrings such as dots or white spaces)
```

```python
# Still need to refer to states by abbreviated names (data dictionary)
abbr_to_state = {
    "AL": "Alabama", "AK": "Alaska", "AZ": "Arizona", "AR": "Arkansas",
    "CA": "California", "CO": "Colorado", "CT": "Connecticut", "DE": "Delaware",
    "FL": "Florida", "GA": "Georgia", "HI": "Hawaii", "ID": "Idaho",
    "IL": "Illinois", "IN": "Indiana", "IA": "Iowa", "KS": "Kansas",
    "KY": "Kentucky", "LA": "Louisiana", "ME": "Maine", "MD": "Maryland",
    "MA": "Massachusetts", "MI": "Michigan", "MN": "Minnesota", "MS": "Mississippi",
    "MO": "Missouri", "MT": "Montana", "NE": "Nebraska", "NV": "Nevada",
    "NH": "New Hampshire", "NJ": "New Jersey", "NM": "New Mexico", "NY": "New York",
    "NC": "North Carolina", "ND": "North Dakota", "OH": "Ohio", "OK": "Oklahoma",
    "OR": "Oregon", "PA": "Pennsylvania", "RI": "Rhode Island", "SC": "South Carolina",
    "SD": "South Dakota", "TN": "Tennessee", "TX": "Texas", "UT": "Utah",
    "VT": "Vermont", "VA": "Virginia", "WA": "Washington", "WV": "West Virginia",
    "WI": "Wisconsin", "WY": "Wyoming", "AS": "American Samoa"
}
```

```python
#creating a new column to dataframe clled primarykey
df['PrimaryKey'] = df['state'].map(abbr_to_state)
```

```python
# if you scroll all the way to the right it shows the full state name as the primary key which matches to your raw data
df.head()
```

```python
# shape/dimension = how many (rows, columns) we have
df.shape
```

```python
states.shape
```

```python
states.head() #we want to merge these coumns into the dataframe (adding 5 columns)
```

```python
# df goes on left, states go on right, identify primary key, identify primary key from other df (state)
# left merge = keep the number of rows/observations from first set on left (how)
first_merge = pd.merge(df, states, left_on='PrimaryKey', right_on='state', how = 'left')
```

```python
first_merge.shape
#rows remained the same which is good because we matched all data from covid19 dataset
#
```

```python
first_merge.columns.tolist() #see what columns we have in the new merged dataset
```

```python
outer_merge = pd.merge(df, states, left_on='PrimaryKey', right_on='state', how = 'outer') #outer merge gives biggest possible merge with values from both datasets
# aka will give values for regions of US in addition to states
```

```python
outer_merge.shape #bigger dataset
```

```python
# find rows where primary key column is empty in this new dataframe
# isna = "is N/A" to find N/A values
outer_merge[outer_merge.PrimaryKey.isna()] #we find out this was not the right merge since we did not match a primary key for DC
```
